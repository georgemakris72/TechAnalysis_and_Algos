```python
import matplotlib.dates as dates
```

```python
idx=mcdon.loc['2007-01-01':'2007-05-01'].index
```

```python
idx
```

```python
stock=mcdon.loc['2007-01-01':'2007-05-01']['Adj. Close']
```

```python
stock.head()
```

```python
fig,ax=plt.subplots()
ax.plot_date(idx,stock,'-')
#ax.yaxis.grid(True)

#locating
ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('%b-%Y'))#sets ticks to beginning of every month

fig.autofmt_xdate()
```

```python
fig,ax=plt.subplots()
ax.plot_date(idx,stock,'-')
#ax.yaxis.grid(True)

#locating
ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_minor_locator(dates.WeekdayLocator(byweekday=0))#0 for Monday
#formatting for major and minor
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n\n\n%b-%Y'))#sets ticks to beginning of every month
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))#sets ticks to beginning of every month



fig.autofmt_xdate()
```
